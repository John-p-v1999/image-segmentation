```
benign_200 = cv2.imread('200x_Benign.png')
malig_200 = cv2.imread('200x_malig.png')
benign_400 = cv2.imread('400x_Benign.png')
malig_400 = cv2.imread('400x_malig.png')
```

```
gray = cv2.cvtColor(benign_200,cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray,128,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh, cmap='gray')
```

```
gray = cv2.cvtColor(benign_400,cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray,120,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh, cmap='gray')
```

```
gray = cv2.cvtColor(malig_200,cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray,110,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh, cmap='gray')
```

```
gray = cv2.cvtColor(malig_400,cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray,135,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh, cmap='gray')
```
